```python
import numpy as np
import hyperobjects as ho
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

np.random.seed(40)
n, m = 3, 50
N = np.random.uniform(-1,1, (n,m))
C = ho.Cube(m, 2)
S = ho.Sphere(m)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1, projection="3d")
(N@C).plot(ax, tol=1, alpha=0.3, color="red")
(ho.radius_grunbaum(m) * N@S).plot(ax, alpha=0.2, color="blue")

norm_N = np.linalg.norm(N, ord=np.inf)
normm_N = np.linalg.norm(N, ord=-np.inf)
norm2_N = np.linalg.norm(N, ord=2)
s = np.linalg.svd(N, compute_uv=False)
print(s)
print(norm2_N)
print(norm_N)
((norm_N/norm2_N) * N@S).plot(ax, alpha=0.2, color="black")
# ((normm_N)*0.5* N@S).plot(ax, alpha=0.2, color="black")
# (norm_N/s.max() * N@S).plot(ax, alpha=0.2, color="green")

# ax.set_xlim(-10,10)
# ax.set_ylim(-10,10)
# ax.set_zlim(-10,10)
```
