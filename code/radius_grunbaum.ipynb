```python
def volume_p_ball(n, r, p):
    if p != 2:
        if p == 1:
            return r**n * 2**n / factorial(n)
        elif p == np.inf:
            return (2*r)**n
        else:
            return r**n * (2*gamma(1/p + 1))**n / gamma(n/p + 1)
    if n==0:
        return 1
    elif n ==1:
        return 2*r
    else:
        return (2/n)*np.pi * r**2 * volume_p_ball(n-2, r, p)
    
def radius_n_sphere(n, V):
    return (1/np.sqrt(np.pi)) * (gamma(n/2 + 1))**(1/n) * V**(1/n)

def volume_n_sphere(n, r):
    return volume_p_ball(n, r, 2)

# X = list(range(2, 100))

# vol_sphere = np.array([volume_n_sphere(n, 1) for n in X]).astype(float)
# vol_demi_sphere = np.array([volume_n_sphere(n, 1/2) for n in X]).astype(float)
# vol_sphere_in_tax = np.array([volume_n_sphere(n, n**(-1/2)) for n in X]).astype(float)
# vol_root_sphere = np.array([volume_n_sphere(n, (np.sqrt(2))) for n in X]).astype(float)
# radius_grunbaum = np.array([ho.radius_grunbaum(n) for n in X]).astype(float)
# vol_grunbaum = np.array([volume_n_sphere(n, ho.radius_grunbaum(n)) for n in X]).astype(float)
# vol_demi_grunbaum = np.array([volume_n_sphere(n, ho.radius_grunbaum(n)/2) for n in X]).astype(float)
# vol_tax_norm = np.array([volume_p_ball(n, 1, 1) for n in X]).astype(float)
# vol_max_norm = np.array([volume_p_ball(n, 1, np.inf) for n in X]).astype(float)

# A_max = np.power(vol_max_norm/vol_sphere, [1/n for n in X])
# A_max_by_2 = np.power(vol_max_norm/vol_root_sphere, [1/n for n in X])
# A_tax = np.power(vol_tax_norm/vol_sphere_in_tax, [1/n for n in X])
# # print(A_tax.round(5).tolist())
# # print(A_max.round(5).tolist())

# k = 2
# C_max = np.array([(4*np.pi*A_max[i])**(n/n-k) for i, n in enumerate(X)])
# C_max_by_2 = np.array([(4*np.pi*A_max_by_2[i])**(n/n-k) for i, n in enumerate(X)])
# C_tax = np.array([(4*np.pi*A_tax[i])**(n/n-k) for i, n in enumerate(X)])
# # print((1/C_max).round(5).tolist())
# # print((1/C_max_by_2).round(5).tolist())
# # print((C_max_by_2).round(5).tolist())
# print(C_tax.round(2).tolist())
# print((1/C_tax).round(2).tolist())
# C = (2*np.exp(0)/np.pi)**(1/2)
# print(C)
# # print(volume_n_sphere(3, C_tax[2]))
# print(volume_n_sphere(3, C))
# print(volume_n_sphere(3, ho.radius_grunbaum(3)/2))
# print(radius_grunbaum.round(5).tolist())
# print((1/radius_grunbaum).round(5).tolist())
# # print((radius_grunbaum-A_tax).round(5).tolist())
```

```python
import itertools
from scipy.linalg import det

n, m = 2, 10
np.random.seed(50)


N = np.random.uniform(-5, 1, size=(n, m))

def zonotope_volume(matrix):
  """Calculates the volume of a zonotope defined by the columns of a matrix.

  Args:
    matrix: A numpy array of shape (n, m) representing the zonotope's generators.

  Returns:
    The volume of the zonotope.
  """

  n, m = matrix.shape
  if n > m:
    raise ValueError("Number of rows must be less than or equal to the number of columns")

  volume = 0
  for indices in itertools.combinations(range(m), n):
    submatrix = matrix[:, indices]
    volume += np.abs(det(submatrix))

  return volume

vol_cube = volume_p_ball(m, 1, p=np.inf)
# vol_cube = 2**m / factorial(m)
vol_zono = zonotope_volume(N)
norm_N = np.linalg.norm(N, ord=2)
norm_N_inf = np.linalg.norm(N, ord=np.inf)
print(norm_N)
print(norm_N_inf)
C = vol_zono / (vol_cube)

r = (1/np.sqrt(np.pi)) * (C/norm_N * gamma(n/2+1)/gamma(m/2 + 1))**(1/(n-m))
# r = 
print("r=", r)
print("Grunbaum:", ho.radius_grunbaum(m))

print("norm_N =",norm_N)
s = np.linalg.svd(N, compute_uv=False)

print(vol_zono / (norm_N * vol_cube))
print((volume_n_sphere(n, r)) / volume_n_sphere(m, r))

print(vol_zono / (norm_N * volume_n_sphere(n, r)))
print(vol_cube / volume_n_sphere(m, r))

r = 1/np.sqrt(np.pi) * ((vol_zono / (norm_N)) * (gamma(m/2 + 1)))**(1/m)
print(r, 2*r)

print(vol_zono, norm_N * volume_n_sphere(m, r))
```

```python
import scipy
import scipy.special

def random_projection_matrix(n):
  """
  Generates a random n x n projection matrix.

  Args:
    n: Dimension of the matrix.

  Returns:
    A numpy array representing the projection matrix.
  """

  # Generate a random orthogonal matrix using QR decomposition
  A = np.random.randn(n, n)
  Q, _ = np.linalg.qr(A)

  # Create a random diagonal matrix with 0s and 1s
  Lambda = np.diag(np.random.randint(0, 2, n))

  # Create the projection matrix
  P = np.dot(np.dot(Q, Lambda), Q.T)

  return P

def generate_n_octahedron_vertices(n):
  """Generates all vertices of an n-dimensional octahedron.

  Args:
    n: Dimensionality of the octahedron.

  Returns:
    A numpy array of shape (2*n, n) containing the vertices.
  """

  I = np.identity(n)
  return np.hstack([I, -I])

n = 3
m = 10

all_norm_proj = []
for i in range(100):
    np.random.seed(i)
    N = np.random.uniform(-1,10000,size=(n,m))
    N = ho.AffineSpace(N.T)
    N = N.orthogonal_projection_matrix
    
    # vertices = N @ (1*generate_n_octahedron_vertices(m))
    # norm_proj = np.linalg.norm(vertices, axis=0, ord=1).max()
    vertices = N @ ho.sample_sphere(m, 2000)
    norm_proj = np.linalg.norm(vertices, axis=0, ord=np.inf).max()
    
    all_norm_proj.append(norm_proj)
    # print(norm_proj)

# print("d", np.sqrt(2))
print("Do we have that the minimum of all these norms is above", ho.radius_grunbaum(n), ho.radius_grunbaum(m))
print(np.max(all_norm_proj))

projec_ln = 2**(1-n) * n * scipy.special.comb(n-1, (n-1)/2)
projec_lm = 2**(1-m) * m * scipy.special.comb(m-1, (m-1)/2)
# print(2*projec_ln, projec_lm)
print(ho.radius_grunbaum(n))
print(np.sqrt(n))
```
