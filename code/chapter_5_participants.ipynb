```python
import numpy as np
import matplotlib.pyplot as plt

sizes_male = [178, 183, 185, 173, 175, 192, 187]
sizes_female = [165, 167, 168]

mass_male = [60.5, 74.5, 84.2, 82.9, 61.8, 87, 77.6]
mass_female = [58.6, 69.2, 73.1]

IMC_male = np.array(mass_male) / (np.array(sizes_male)*0.01)**2
IMC_female = np.array(mass_female) / (np.array(sizes_female)*0.01)**2

nb_sport_per_week = [6,4,4,5,0,6, 5.33,0, 3, 10]
```

```python
%matplotlib inline

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111) #, projection='3d')

ax.boxplot([sizes_male,sizes_female])

ax.set_title(r"$m = 10$")

fig.tight_layout()
plt.show()
```

```python
%matplotlib inline

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111) #, projection='3d')

ax.boxplot([mass_male, mass_female])
ax.set_title(r"$m = 10$")

fig.tight_layout()
plt.show()
```

```python
%matplotlib inline

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111) #, projection='3d')

ax.boxplot([IMC_male, IMC_female])
ax.set_title(r"IMC")

fig.tight_layout()
plt.show()
```

```python
%matplotlib inline

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111) #, projection='3d')

ax.boxplot(nb_sport_per_week)
ax.set_title(r"$m = 10$")

fig.tight_layout()
plt.show()
```
