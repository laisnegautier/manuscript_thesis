```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection, Line3DCollection
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)

def f_A(a, l, lo):
    return a * np.exp( - (l/lo - 1)**2 / 0.45)

def f_P(l, lo, eps_o = 0.6):
    return (np.exp(5 * (l/lo - 1) / eps_o) - 1) / (np.exp(5) - 1)

lengths = np.linspace(0.0, 0.20, 100)
ax.plot(lengths, [f_A(1, l, 0.068) for l in lengths], c="orange", label="Supraspinatus")
ax.plot(lengths, [f_P(l, 0.068) for l in lengths], c="orange")

ax.plot(lengths, [f_A(1, l, 0.098) for l in lengths], c="blue", label="Anterior Deltoid")
ax.plot(lengths, [f_P(l, 0.098) for l in lengths], c="blue")

ax.plot(lengths, [f_A(1, l, 0.134) for l in lengths], c="green", label="Triceps long")
ax.plot(lengths, [f_P(l, 0.134) for l in lengths], c="green")
ax.plot(lengths, [f_P(l, 0.134, 0.5) for l in lengths], c="violet")

ax.set_xlabel(r'Muscle length (meters)')
plt.legend()
# print(lengths)

ax.set_xticklabels([0.0, 0.01, 2])

ax.set_xlim(0, 0.20)
ax.set_ylim(0, 1.2)
# plt.savefig("chapter4.pdf", format="pdf", bbox_inches="tight")
```
