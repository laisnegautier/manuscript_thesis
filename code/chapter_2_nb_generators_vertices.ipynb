```python
import numpy as np
import matplotlib.pyplot as plt
import common
import math
# plt.rcParams['text.usetex'] = True
```

```python
%matplotlib inline

fig = plt.figure(figsize=(10,5))
# fig, ax = plt.subplots(1,3)

ax = fig.add_subplot(111)
np.random.seed(40)

ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
# ax.set_zlabel(r'$z$')

n = np.arange(1, 10)
m = 50
x = n
y = [common.nb_k_face_zonotope(0, i, m) for i in n]
ax.scatter(x, y)
ax.plot(x, y)

# Loop for annotation of all points 
text = [f"{int(common.nb_k_face_zonotope(0, i, m))}" for i in n]
text[4] = r'$\approx 5\times 10^{5}$'
text[5] = r'$\approx 4\times 10^{6}$'
text[6] = r'$\approx 3\times 10^{7}$'
text[7] = r'$\approx 2\times 10^{8}$'
for i in range(len(x)): 
    plt.annotate(text[i], (x[i], y[i] + 50000000.4)) 

# ax.set_xlim(-10,10)
# ax.set_ylim(-10,10)
# ax.set_zlim(-1000,1000)

fig.tight_layout()
plt.show()
```
